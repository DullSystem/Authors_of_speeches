## Шаг 7. Общий вывод

**1.** Количество конференции резко увеличилось с 2008 по 2009 года с 15 до 226, самый первый скачок наблюдается с 2001 года по 2002 год, количество конференция увеличилось на 22.
- Всего 14 полей и 3456 строк

**2.** Можно увидеть, что больше всего выступлений проходило в **США - 50,75% , Канаде - 19,35% и Великобритании - 16,68%**
- Можно выделить наиболее популярные категории по абсолютным числам: **технологии, общество, наука, глобальные проблемы и культура**
- С 2020 по 2021 года состав уменьшился, популярные категории: технологии, общество и другие.

**3.** Наиболее популярные категории выступления: наука, технологии, общество и другие.
- Наука, технологии и культура популярны во всех 3 странах. 
- Чаще всего аплодисменты вызывают развлечения и социальные изменения, а реже всего глобальные проблемы и наука.
- Категории образование и бизнес чаще всего вызвают смех, глобальные проблемы реже.
- Зависимости между длительностью выступления и количеством просмотров нет. 
- Чаще всего смотрели **Do schools kill creativity?** длительностью 19,40 минут и с просмотрами 72 млн,
- Самое долгое выступлени длилось 62,58 минут, название **3 secrets to Netflixs success**, количество просмотров 1 847 803 млн.

**4.** Самые основные категории деятельности авторов это: писатель, журналист и художник.
- Чаще всего 1 выступление, больше всех выступал **HansRosling**
- Из дизайнеров чаще всех выступал **TomWujec**, самое малое количество просмотров 2 выступления: **3 ways the brain creates meaning и Learn to use the 13th-centure astrolabe.**

**5.** Количество конференций в 2019-2021 годах уменьшилось, наиболее популярые категории **perfomance и art**, самые не популярные **politics и spase**.
- Популярная конференция **TED2019**, было 7 категорий из которых категория art заняла 4 выступления, **Ted Residency** с одним выступлением занимает последнее место.
